이상치 처리

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 모든 int64형 데이터에 대한 반복문
for column in df.select_dtypes(include='int64').columns:
    # 히스토그램 그리기
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')

    # 박스 플롯 그리기
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[column], color='skyblue')
    plt.title(f'Boxplot of {column}')

    plt.show()
```

```python
# 박스 플롯 그리기
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.select_dtypes(include='int64'), color='skyblue')
plt.title('Boxplot of Integer Columns')
plt.xticks(rotation=45)  # x축 레이블 회전
plt.tight_layout()  # 레이아웃 조정
plt.show()
```

```python
# 시골 통학거리 시각화
df['address']
df['traveltime']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 주소(address)별 이동 시간(traveltime)의 분포를 확인하는 박스플롯
plt.figure(figsize=(10, 6))
sns.boxplot(x='address', y='traveltime', data=df)
plt.title('Boxplot of Travel Time by Address')
plt.xlabel('Address')
plt.ylabel('Travel Time')
plt.show()

# 주소(address)별 이동 시간(traveltime)의 히스토그램
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='traveltime', hue='address', bins=10, kde=True)
plt.title('Histogram of Travel Time by Address')
plt.xlabel('Travel Time')
plt.ylabel('Frequency')
plt.show()

# 주소(address)별 이동 시간(traveltime)의 평균 비교
mean_traveltime_by_address = df.groupby('address')['traveltime'].mean()
plt.figure(figsize=(8, 5))
mean_traveltime_by_address.plot(kind='bar', color='skyblue')
plt.title('Mean Travel Time by Address')
plt.xlabel('Address')
plt.ylabel('Mean Travel Time')
plt.xticks(rotation=0)
plt.show()
```
